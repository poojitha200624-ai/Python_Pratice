```python
#Q1. Price After Discount
price = float(input("Enter price: "))
discount = float(input("Enter discount percentage: "))
final_price = price - (price * discount / 100)
print("Price after discount =", final_price)
```

```python
#Q2. Convert Minutes to Seconds
minutes = int(input("Enter minutes: "))
seconds = minutes * 60
print("Seconds =", seconds)
```

```python
#Q3. Split the Bill Among 3 Friends
bill = float(input("Enter total bill amount: "))
share = bill / 3
print("Each friend pays =", share)
```

```python
#Calculate Distance Travelled
speed = float(input("Enter speed: "))
time = float(input("Enter time: "))
distance = speed * time
print("Distance travelled =", distance)
```

```python
#. Convert Kilometers to Meters
km = float(input("Enter kilometers: "))
meters = km * 1000
print("Meters =", meters)
```

```python
#Q6. Convert Bytes to Bits
bytes_value = int(input("Enter bytes: "))
bits = bytes_value * 8
print("Bits =", bits)
```

```python
#Q7. Distribute Candies Equally Among People
candies = int(input("Enter total candies: "))
people = int(input("Enter number of people: "))
each = candies // people
remaining = candies % people
print("Each person gets =", each)
print("Remaining candies =", remaining)
```

```python
#Q9. Calculate Simple Interest
cakes_per_day = int(input("Enter cakes made per day: "))
days = int(input("Enter number of days: "))
total = cakes_per_day * days
print("Total cakes made =", total)
```

```python
#Q10. Calculate Salary of Employee
basic = float(input("Enter basic salary: "))
hra = float(input("Enter HRA: "))
da = float(input("Enter DA: "))
salary = basic + hra + da
print("Total Salary =", salary)
```

```python
#Q11. Selling Price (given Cost Price and Profit)
cp = float(input("Enter cost price: "))
profit = float(input("Enter profit: "))
sp = cp + profit
print("Selling Price =", sp)
```

```python
#Q12. Cost Price (given Selling Price and Loss)
sp = float(input("Enter selling price: "))
loss = float(input("Enter loss: "))
cp = sp + loss
print("Cost Price =", cp)
```

```python
#Q14. Pens Discount (Buy 5, Pay for 3)
quantity = int(input("Enter number of pens: "))
price = float(input("Enter price per pen: "))
free_sets = quantity // 5
pay_for = quantity - (free_sets * 2)
total = pay_for * price
print("Total amount to pay =", total)
```

```python
#Q13. Mangoes Discount (Buy 3 Get 1 Free)
quantity = int(input("Enter number of mangoes: "))
price = float(input("Enter price per mango: "))
pay_for = quantity - (quantity // 4)
total = pay_for * price
print("Total amount to pay =", total)
```
