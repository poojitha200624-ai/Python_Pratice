```python
n = int(input("Enter number: "))
while n >= 10:
    n = n // 10
if n % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
n = int(input("Enter number: "))
d = int(input("Enter digit to search: "))
found = False
while n > 0:
    rem = n % 10
    if rem == d:
        found = True
        break
    n = n // 10
if found:
    print("Digit found")
else:
    print("Digit not found")
```

```python
n = int(input("Enter number: "))
d = int(input("Enter digit: "))
count = 0
while n > 0:
    rem = n % 10
    if rem == d:
        count += 1
    n = n // 10
print("Count:", count)
```

```python
n = int(input("Enter number: "))
temp = n
prev = 10
flag = True
while temp > 0:
    rem = temp % 10
    if rem > prev:
        flag = False
        break
    prev = rem
    temp = temp // 10
if flag:
    print("Ascending")
else:
    print("Not Ascending")
```

```python
n = int(input("Enter number: "))
temp = n
prev = -1
flag = True
while temp > 0:
    rem = temp % 10
    if rem < prev:
        flag = False
        break
    prev = rem
    temp = temp // 10
if flag:
    print("Descending")
else:
    print("Not Descending")
```

```python
n = int(input("Enter number: "))
temp = n
last = n % 10
count = 0
while temp > 0:
    first = temp % 10
    temp = temp // 10
    count += 1
power = 10 ** (count - 1)
middle = (n % power) // 10
new_num = last * power + middle * 10 + first
print("After swap:", new_num)
```

```python
n = input("Enter number: ")
for i in range(10):
    count = 0
    for j in n:
        if int(j) == i:
            count += 1
    if count > 0:
        print(i, ":", count)
```

```python
n = int(input("Enter number: "))
largest = 0
while n > 0:
    rem = n % 10
    if rem > largest:
        largest = rem
    n = n // 10
print("Largest digit:", largest)
```

```python
n = int(input("Enter number: "))
smallest = 9
while n > 0:
    rem = n % 10
    if rem < smallest:
        smallest = rem
    n = n // 10
print("Smallest digit:", smallest)
```

```python
n = int(input("Enter number: "))
even = 0
odd = 0
while n > 0:
    rem = n % 10
    if rem % 2 == 0:
        even += 1
    else:
        odd += 1
    n = n // 10
if even > 0 and odd == 0:
    print("All Even")
elif odd > 0 and even == 0:
    print("All Odd")
else:
    print("Mixed")
```

```python
n = int(input("Enter number: "))
even_sum = 0
odd_sum = 0
while n > 0:
    rem = n % 10
    if rem % 2 == 0:
        even_sum += rem
    else:
        odd_sum += rem
    n = n // 10
print("Difference:", even_sum - odd_sum)
```

```python
n = input("Enter number: ")
result = ""
for i in n:
    if i != '0':
        result += i
print("After removing zeros:", result)
```

```python
n = int(input("Enter number: "))
sq = 0
cube = 0
while n > 0:
    rem = n % 10
    sq += rem ** 2
    cube += rem ** 3
    n = n // 10
print("Sum of squares:", sq)
print("Sum of cubes:", cube)
```
